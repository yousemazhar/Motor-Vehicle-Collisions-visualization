```python
# Create copies to preserve original data
df_crashes_clean = df_crashes.copy()
df_persons_clean = df_persons.copy()

print("=== HANDLING MISSING VALUES ===\n")
```

```python
# ========== CRASHES DATASET ==========

# 1. Drop columns with >90% missing (VEHICLE/CONTRIBUTING FACTOR 3, 4, 5)
# These columns represent 3rd, 4th, and 5th vehicles which are rarely involved
columns_to_drop_crashes = [
    'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5',
    'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5'
]

df_crashes_clean = df_crashes_clean.drop(columns=columns_to_drop_crashes)
print(f" Dropped {len(columns_to_drop_crashes)} columns with >90% missing values")

# 2. Handle OFF STREET NAME & CROSS STREET NAME (82% and 38% missing)
# These are location descriptors - keep as is since they're mutually exclusive with ON STREET NAME
# Missing is informative (crash occurred on a street vs off street)
print(" Kept OFF STREET NAME and CROSS STREET NAME (missing is meaningful)")

# 3. Handle BOROUGH and ZIP CODE (~31% missing)
# Fill with 'Unknown' as these are categorical location data
df_crashes_clean['BOROUGH'] = df_crashes_clean['BOROUGH'].fillna('Unknown')
df_crashes_clean['ZIP CODE'] = df_crashes_clean['ZIP CODE'].fillna('Unknown')
print(" Filled BOROUGH and ZIP CODE with 'Unknown'")

# 4. Handle ON STREET NAME (~22% missing)
# Keep as is - missing when crash is OFF STREET
print(" Kept ON STREET NAME (missing is meaningful)")

# 5. Handle VEHICLE TYPE CODE 2 and CONTRIBUTING FACTOR VEHICLE 2 (~20% and 16% missing)
# Fill with 'No Second Vehicle' since missing indicates single-vehicle crash
df_crashes_clean['VEHICLE TYPE CODE 2'] = df_crashes_clean['VEHICLE TYPE CODE 2'].fillna('No Second Vehicle')
df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 2'] = df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 2'].fillna('No Second Vehicle')
print(" Filled VEHICLE 2 columns with 'No Second Vehicle'")

# 6. Handle LOCATION, LATITUDE, LONGITUDE (~11% missing)
# Drop rows with missing coordinates as they're critical for spatial analysis
initial_crashes_count = len(df_crashes_clean)
df_crashes_clean = df_crashes_clean.dropna(subset=['LATITUDE', 'LONGITUDE'])
dropped_crashes = initial_crashes_count - len(df_crashes_clean)
print(f" Dropped {dropped_crashes} rows with missing LATITUDE/LONGITUDE")

# Also drop the LOCATION column (redundant with LAT/LONG)
df_crashes_clean = df_crashes_clean.drop(columns=['LOCATION'])
print(" Dropped LOCATION column (redundant)")

# 7. Handle VEHICLE TYPE CODE 1 and CONTRIBUTING FACTOR VEHICLE 1 (<1% missing)
# Fill with 'Unknown' as every crash should have at least one vehicle
df_crashes_clean['VEHICLE TYPE CODE 1'] = df_crashes_clean['VEHICLE TYPE CODE 1'].fillna('Unknown')
df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 1'] = df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 1'].fillna('Unknown')
print(" Filled VEHICLE 1 columns with 'Unknown'")

# 8. Handle NUMBER OF PERSONS INJURED/KILLED (very minimal missing)
# Fill with 0 as missing likely means no casualties
df_crashes_clean['NUMBER OF PERSONS INJURED'] = df_crashes_clean['NUMBER OF PERSONS INJURED'].fillna(0)
df_crashes_clean['NUMBER OF PERSONS KILLED'] = df_crashes_clean['NUMBER OF PERSONS KILLED'].fillna(0)
print(" Filled injury/death counts with 0")

print(f"\n Crashes dataset: {len(df_crashes)}  {len(df_crashes_clean)} rows")
print(f"  Remaining missing values: {df_crashes_clean.isnull().sum().sum()}")
```

```python
# ========== PERSONS DATASET ==========

# 1. Drop columns with >90% missing (CONTRIBUTING_FACTOR_1 & 2, PED columns)
# PED_ACTION, PED_LOCATION, CONTRIBUTING_FACTOR columns are 98%+ missing
columns_to_drop_persons = [
    'CONTRIBUTING_FACTOR_1', 'CONTRIBUTING_FACTOR_2',
    'PED_ACTION', 'PED_LOCATION'
]

df_persons_clean = df_persons_clean.drop(columns=columns_to_drop_persons)
print(f"\n Dropped {len(columns_to_drop_persons)} columns with >98% missing values")

# 2. Handle SAFETY_EQUIPMENT, POSITION_IN_VEHICLE, EJECTION (~48-52% missing)
# These apply only to vehicle occupants, not pedestrians
# Fill with 'Not Applicable'
df_persons_clean['SAFETY_EQUIPMENT'] = df_persons_clean['SAFETY_EQUIPMENT'].fillna('Not Applicable')
df_persons_clean['POSITION_IN_VEHICLE'] = df_persons_clean['POSITION_IN_VEHICLE'].fillna('Not Applicable')
df_persons_clean['EJECTION'] = df_persons_clean['EJECTION'].fillna('Not Applicable')
print(" Filled vehicle-occupant columns with 'Not Applicable'")

# 3. Handle EMOTIONAL_STATUS, BODILY_INJURY, COMPLAINT (~47% missing)
# Fill with 'Not Reported' as these are optional fields
df_persons_clean['EMOTIONAL_STATUS'] = df_persons_clean['EMOTIONAL_STATUS'].fillna('Not Reported')
df_persons_clean['BODILY_INJURY'] = df_persons_clean['BODILY_INJURY'].fillna('Not Reported')
df_persons_clean['COMPLAINT'] = df_persons_clean['COMPLAINT'].fillna('Not Reported')
print(" Filled medical/emotional status with 'Not Reported'")

# 4. Handle PERSON_AGE (~11% missing)
# Fill with median age (36) based on descriptive statistics
df_persons_clean['PERSON_AGE'] = pd.to_numeric(
    df_persons_clean['PERSON_AGE'],
    errors='coerce'
)

median_age = df_persons_clean['PERSON_AGE'].median()
df_persons_clean['PERSON_AGE'] = df_persons_clean['PERSON_AGE'].fillna(median_age)
print(f" Filled PERSON_AGE with median value ({median_age})")

# 5. Handle PERSON_SEX (~11% missing)
# Fill with 'Unknown'
df_persons_clean['PERSON_SEX'] = df_persons_clean['PERSON_SEX'].fillna('Unknown')
print(" Filled PERSON_SEX with 'Unknown'")

# 6. Handle VEHICLE_ID (~4% missing)
# Keep as is - missing for pedestrians who aren't in vehicles
print(" Kept VEHICLE_ID (missing is meaningful for pedestrians)")

# 7. Handle PED_ROLE (~3% missing)
# Fill with 'Unknown' - this should have a value but is missing
df_persons_clean['PED_ROLE'] = df_persons_clean['PED_ROLE'].fillna('Unknown')
print(" Filled PED_ROLE with 'Unknown'")

# 8. Handle PERSON_ID (minimal missing)
# Drop rows with missing PERSON_ID as it's a critical identifier
initial_persons_count = len(df_persons_clean)
df_persons_clean = df_persons_clean.dropna(subset=['PERSON_ID'])
dropped_persons = initial_persons_count - len(df_persons_clean)
print(f" Dropped {dropped_persons} rows with missing PERSON_ID")

print(f"\n Persons dataset: {len(df_persons)}  {len(df_persons_clean)} rows")
print(f"  Remaining missing values: {df_persons_clean.isnull().sum().sum()}")
```

```python
# Final verification
print("\n" + "="*60)
print("FINAL MISSING VALUES CHECK")
print("="*60)

print("\n--- Crashes Dataset ---")
crashes_remaining_missing = df_crashes_clean.isnull().sum()
crashes_remaining_missing = crashes_remaining_missing[crashes_remaining_missing > 0]
if len(crashes_remaining_missing) == 0:
    print(" No missing values remaining!")
else:
    print(crashes_remaining_missing)

print("\n--- Persons Dataset ---")
persons_remaining_missing = df_persons_clean.isnull().sum()
persons_remaining_missing = persons_remaining_missing[persons_remaining_missing > 0]
if len(persons_remaining_missing) == 0:
    print(" No missing values remaining!")
else:
    print(persons_remaining_missing)

print("\n" + "="*60)
print(f" Data cleaning complete!")
print(f"  Crashes: {df_crashes_clean.shape}")
print(f"  Persons: {df_persons_clean.shape}")
print("="*60)
```

```python
print("=== OUTLIER DETECTION - CRASHES DATASET ===\n")

# 1. Check LATITUDE and LONGITUDE for NYC boundaries
# NYC approximate boundaries: Lat (40.4-40.95), Long (-74.3 to -73.7)
print("--- Geospatial Outliers ---")
lat_outliers = df_crashes_clean[(df_crashes_clean['LATITUDE'] < 40.4) |
                                 (df_crashes_clean['LATITUDE'] > 40.95)]
lon_outliers = df_crashes_clean[(df_crashes_clean['LONGITUDE'] < -74.3) |
                                 (df_crashes_clean['LONGITUDE'] > -73.7)]

print(f"Latitude outliers (outside 40.4-40.95): {len(lat_outliers)}")
print(f"Longitude outliers (outside -74.3 to -73.7): {len(lon_outliers)}")

# 2. Check casualty counts for extreme values
print("\n--- Casualty Count Outliers ---")
print("\nInjury/Death Statistics:")
print(df_crashes_clean[['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
                        'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
                        'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
                        'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']].describe())

# Identify extreme casualties using IQR method
Q1_injured = df_crashes_clean['NUMBER OF PERSONS INJURED'].quantile(0.25)
Q3_injured = df_crashes_clean['NUMBER OF PERSONS INJURED'].quantile(0.75)
IQR_injured = Q3_injured - Q1_injured
upper_bound_injured = Q3_injured + 3 * IQR_injured  # Using 3*IQR for extreme outliers

extreme_injuries = df_crashes_clean[df_crashes_clean['NUMBER OF PERSONS INJURED'] > upper_bound_injured]
print(f"\nCrashes with extreme injuries (>{upper_bound_injured}): {len(extreme_injuries)}")

extreme_deaths = df_crashes_clean[df_crashes_clean['NUMBER OF PERSONS KILLED'] > 2]
print(f"Crashes with >2 deaths: {len(extreme_deaths)}")
```

```python
print("\n=== HANDLING OUTLIERS - CRASHES DATASET ===\n")

# 1. Remove geospatial outliers (coordinates outside NYC)
initial_count = len(df_crashes_clean)
df_crashes_clean = df_crashes_clean[
    (df_crashes_clean['LATITUDE'].between(40.4, 40.95)) &
    (df_crashes_clean['LONGITUDE'].between(-74.3, -73.7))
]
removed_geo = initial_count - len(df_crashes_clean)
print(f" Removed {removed_geo} crashes with coordinates outside NYC boundaries")

# 2. Keep extreme casualty counts (they're legitimate, though rare)
print(f" Kept extreme casualty counts (legitimate rare events)")

# 3. Check for impossible negative values or zeros in required fields
print("\n--- Checking for Invalid Values ---")
invalid_collision_ids = df_crashes_clean[df_crashes_clean['COLLISION_ID'] <= 0]
print(f"Invalid COLLISION_IDs: {len(invalid_collision_ids)}")

print(f"\n Final crashes dataset size: {len(df_crashes_clean)} rows")
```

```python
print("\n=== STANDARDIZING FORMATS ===\n")

# ========== CRASHES DATASET ==========
print("--- Crashes Dataset ---")

# 1. Convert CRASH DATE to datetime
df_crashes_clean['CRASH DATE'] = pd.to_datetime(df_crashes_clean['CRASH DATE'], format='%m/%d/%Y')
print(" Converted CRASH DATE to datetime format")

# 2. Standardize CRASH TIME format (already in HH:MM format, verify)
print(f" CRASH TIME format verified (HH:MM)")

# 3. Extract useful date components
df_crashes_clean['CRASH_YEAR'] = df_crashes_clean['CRASH DATE'].dt.year
df_crashes_clean['CRASH_MONTH'] = df_crashes_clean['CRASH DATE'].dt.month
df_crashes_clean['CRASH_DAY'] = df_crashes_clean['CRASH DATE'].dt.day
df_crashes_clean['CRASH_DAYOFWEEK'] = df_crashes_clean['CRASH DATE'].dt.dayofweek  # 0=Monday, 6=Sunday
df_crashes_clean['CRASH_HOUR'] = pd.to_datetime(df_crashes_clean['CRASH TIME'], format='%H:%M').dt.hour
print(" Extracted date/time components (year, month, day, dayofweek, hour)")

# ========== PERSONS DATASET ==========
print("\n--- Persons Dataset ---")

# 1. Convert CRASH_DATE to datetime
df_persons_clean['CRASH_DATE'] = pd.to_datetime(df_persons_clean['CRASH_DATE'], format='%m/%d/%Y')
print(" Converted CRASH_DATE to datetime format")

# 2. Verify CRASH_TIME format
print(f" CRASH_TIME format verified (HH:MM)")
```

```python
# ========== CRASHES DATASET ==========
print("\n--- String Standardization: Crashes Dataset ---")

# 1. Standardize categorical text columns (uppercase, strip whitespace)
text_columns_crashes = ['BOROUGH', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME',
                        'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2',
                        'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']

for col in text_columns_crashes:
    if col in df_crashes_clean.columns:
        df_crashes_clean[col] = df_crashes_clean[col].astype(str).str.strip().str.upper()

print(f" Standardized {len(text_columns_crashes)} text columns (uppercase, trimmed)")

# 2. Standardize ZIP CODE format (ensure consistent length)
df_crashes_clean['ZIP CODE'] = df_crashes_clean['ZIP CODE'].astype(str).str.strip()
print(" Standardized ZIP CODE format")

# 3. Check unique values in key categorical columns
print("\n--- Key Categorical Values ---")
print(f"Unique BOROUGH values: {df_crashes_clean['BOROUGH'].unique()}")
print(f"Total unique CONTRIBUTING FACTOR VEHICLE 1: {df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 1'].nunique()}")
print(f"Total unique VEHICLE TYPE CODE 1: {df_crashes_clean['VEHICLE TYPE CODE 1'].nunique()}")

# ========== PERSONS DATASET ==========
print("\n--- String Standardization: Persons Dataset ---")

# 1. Standardize categorical text columns
text_columns_persons = ['PERSON_TYPE', 'PERSON_INJURY', 'EJECTION', 'EMOTIONAL_STATUS',
                        'BODILY_INJURY', 'POSITION_IN_VEHICLE', 'SAFETY_EQUIPMENT',
                        'COMPLAINT', 'PED_ROLE', 'PERSON_SEX']

for col in text_columns_persons:
    if col in df_persons_clean.columns:
        df_persons_clean[col] = df_persons_clean[col].astype(str).str.strip().str.upper()

print(f" Standardized {len(text_columns_persons)} text columns (uppercase, trimmed)")

# 2. Check unique values in key categorical columns
print("\n--- Key Categorical Values ---")
print(f"Unique PERSON_TYPE values: {df_persons_clean['PERSON_TYPE'].unique()}")
print(f"Unique PERSON_INJURY values: {df_persons_clean['PERSON_INJURY'].unique()}")
print(f"Unique PERSON_SEX values: {df_persons_clean['PERSON_SEX'].unique()}")
```

```python
print("\n--- Category Consolidation ---")

# ========== CRASHES DATASET ==========

# 1. Clean up BOROUGH values
df_crashes_clean['BOROUGH'] = df_crashes_clean['BOROUGH'].replace({
    'NAN': 'UNKNOWN',
    'NONE': 'UNKNOWN'
})
print(" Consolidated BOROUGH categories")

# 2. Group similar contributing factors (optional - show top factors)
print("\nTop 10 Contributing Factors:")
print(df_crashes_clean['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(10))

# 3. Group similar vehicle types (optional - show top types)
print("\nTop 10 Vehicle Types:")
print(df_crashes_clean['VEHICLE TYPE CODE 1'].value_counts().head(10))

# ========== PERSONS DATASET ==========

# 1. Standardize PERSON_SEX values
df_persons_clean['PERSON_SEX'] = df_persons_clean['PERSON_SEX'].replace({
    'U': 'UNKNOWN',
    'NAN': 'UNKNOWN'
})
print("\n Consolidated PERSON_SEX categories")

# 2. Show person type distribution
print("\nPerson Type Distribution:")
print(df_persons_clean['PERSON_TYPE'].value_counts())
```

```python
print("\n=== PREPARING DATASETS FOR INTEGRATION ===\n")

# Identify and handle redundant columns before joining
print("--- Handling Redundant Columns ---")

# Rename conflicting columns in persons dataset to avoid duplication
df_persons_prepared = df_persons_clean.copy()

# Drop redundant columns from persons dataset (already in crashes)
redundant_cols = ['CRASH_DATE', 'CRASH_TIME']
df_persons_prepared = df_persons_prepared.drop(columns=redundant_cols)
print(f" Dropped redundant columns from persons dataset: {redundant_cols}")

# Rename UNIQUE_ID to PERSON_UNIQUE_ID for clarity
df_persons_prepared = df_persons_prepared.rename(columns={'UNIQUE_ID': 'PERSON_UNIQUE_ID'})
print(" Renamed UNIQUE_ID to PERSON_UNIQUE_ID")

print(f"\nCrashes dataset columns: {len(df_crashes_clean.columns)}")
print(f"Persons dataset columns (prepared): {len(df_persons_prepared.columns)}")
```

```python
print("\n=== PERFORMING DATA INTEGRATION ===\n")

# Perform LEFT JOIN: Keep all crashes, add person-level details
# This creates one row per person involved in each crash
df_integrated = df_crashes_clean.merge(
    df_persons_prepared,
    on='COLLISION_ID',
    how='left',
    suffixes=('_CRASH', '_PERSON')
)

print(f" Join completed!")
print(f"  Crashes dataset: {len(df_crashes_clean)} rows")
print(f"  Persons dataset: {len(df_persons_prepared)} rows")
print(f"  Integrated dataset: {len(df_integrated)} rows")
print(f"  Integrated dataset columns: {len(df_integrated.columns)}")
```

```python
print("\n=== INTEGRATED DATASET INSPECTION ===\n")

# Display basic info
print("--- Shape and Structure ---")
print(f"Shape: {df_integrated.shape}")
print(f"\nFirst few rows:")
display(df_integrated.head())

print("\n--- Data Types ---")
print(df_integrated.dtypes)
```

```python
print("\n=== RESOLVING MISSING VALUES FROM JOIN ===\n")

# Check for new missing values
print("--- Missing Values Analysis ---")
integrated_missing = df_integrated.isnull().sum()
integrated_missing_pct = (integrated_missing / len(df_integrated)) * 100
missing_summary = pd.DataFrame({
    'Column': integrated_missing.index,
    'Missing_Count': integrated_missing.values,
    'Missing_Percentage': integrated_missing_pct.values
})
missing_summary = missing_summary[missing_summary['Missing_Count'] > 0].sort_values(
    'Missing_Percentage', ascending=False
)

print("\nColumns with missing values:")
display(missing_summary)

# Handle missing values from join
print("\n--- Handling Join-Related Missing Values ---")

# Person-level columns will have missing values for crashes without person records
# These are crashes where person details weren't recorded
person_columns = [col for col in df_integrated.columns if col in df_persons_prepared.columns]

# Fill person-level missing values appropriately
for col in person_columns:
    if col in ['PERSON_UNIQUE_ID', 'PERSON_ID', 'VEHICLE_ID']:
        # Keep as NaN for identifier columns (no person record exists)
        continue
    elif df_integrated[col].dtype == 'object':
        # Fill categorical columns with 'No Person Record'
        df_integrated[col] = df_integrated[col].fillna('No Person Record')
    elif df_integrated[col].dtype in ['float64', 'int64']:
        # Fill numerical columns with appropriate values
        if 'AGE' in col:
            # Use median age for missing
            df_integrated[col] = df_integrated[col].fillna(df_integrated[col].median())
        else:
            # Use 0 for other numerical columns
            df_integrated[col] = df_integrated[col].fillna(0)

print(f" Filled missing values in person-level columns")

# Verify remaining missing values
remaining_missing = df_integrated.isnull().sum()
remaining_missing = remaining_missing[remaining_missing > 0]
print(f"\n--- Remaining Missing Values ---")
if len(remaining_missing) == 0:
    print(" No missing values remaining!")
else:
    print(remaining_missing)
```

```python
print("\n=== RESOLVING COLUMN INCONSISTENCIES ===\n")

# 1. Check for duplicate column names with suffixes
duplicate_suffix_cols = [col for col in df_integrated.columns if '_CRASH' in col or '_PERSON' in col]
print(f"--- Columns with suffixes (potential duplicates): {len(duplicate_suffix_cols)} ---")
if len(duplicate_suffix_cols) > 0:
    print(duplicate_suffix_cols)

# 2. Remove truly redundant columns
# Since we already dropped CRASH_DATE and CRASH_TIME from persons before join,
# we shouldn't have these duplicates

# 3. Check for columns that should be consolidated
print("\n--- Column Consolidation Check ---")
print(f"Total columns: {len(df_integrated.columns)}")

# List all columns for review
print("\nAll columns in integrated dataset:")
for i, col in enumerate(df_integrated.columns, 1):
    print(f"{i}. {col}")
```

```python
print("\n=== RESOLVING DATA TYPE MISMATCHES ===\n")

# Check current data types
print("--- Current Data Types ---")
print(df_integrated.dtypes.value_counts())

# Ensure consistent data types
print("\n--- Standardizing Data Types ---")

# 1. Ensure date columns are datetime
if 'CRASH DATE' in df_integrated.columns:
    df_integrated['CRASH DATE'] = pd.to_datetime(df_integrated['CRASH DATE'])
    print(" CRASH DATE is datetime")

# 2. Ensure integer columns for counts
count_columns = [col for col in df_integrated.columns if 'NUMBER' in col.upper()]
for col in count_columns:
    if col in df_integrated.columns:
        df_integrated[col] = df_integrated[col].fillna(0).astype(int)
print(f" Converted {len(count_columns)} count columns to integer")

# 3. Ensure ID columns are appropriate type
id_columns = ['COLLISION_ID', 'PERSON_UNIQUE_ID']
for col in id_columns:
    if col in df_integrated.columns:
        # Keep COLLISION_ID as int, PERSON_UNIQUE_ID can be float (has NaN)
        if col == 'COLLISION_ID':
            df_integrated[col] = df_integrated[col].astype(int)
            print(f" {col} converted to integer")

# 4. Ensure categorical columns are string type
categorical_cols = ['BOROUGH', 'PERSON_TYPE', 'PERSON_INJURY', 'PERSON_SEX',
                    'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']
for col in categorical_cols:
    if col in df_integrated.columns:
        df_integrated[col] = df_integrated[col].astype(str)
print(f" Standardized categorical columns to string type")

# 5. Ensure age and coordinates are float
numeric_cols = ['PERSON_AGE', 'LATITUDE', 'LONGITUDE']
for col in numeric_cols:
    if col in df_integrated.columns:
        df_integrated[col] = df_integrated[col].astype(float)
print(f" Standardized numeric columns to float type")

print("\n--- Updated Data Types ---")
print(df_integrated.dtypes.value_counts())
```

```python
print("\n=== FINAL VALIDATION ===\n")

# Check for any remaining issues
print("--- Data Quality Checks ---")
print(f"1. Total rows: {len(df_integrated):,}")
print(f"2. Total columns: {len(df_integrated.columns)}")
print(f"3. Duplicate rows: {df_integrated.duplicated().sum()}")
print(f"4. Missing values: {df_integrated.isnull().sum().sum()}")
print(f"5. Memory usage: {df_integrated.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Check collision coverage
print("\n--- Join Coverage ---")
print(f"Unique crashes in integrated dataset: {df_integrated['COLLISION_ID'].nunique()}")
print(f"Crashes with person records: {df_integrated[df_integrated['PERSON_UNIQUE_ID'].notna()]['COLLISION_ID'].nunique()}")
print(f"Crashes without person records: {df_integrated[df_integrated['PERSON_UNIQUE_ID'].isna()]['COLLISION_ID'].nunique()}")

# Summary statistics
print("\n--- Summary Statistics ---")
print(df_integrated.describe())
```
